```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next = None
        self.prev= None
        
class DoublyLL:
    def __init__(self):
        self.head=None
        self.tail=None
    
    def __iter__(self):
        node=self.head
        while node:
            yield node
            node = node.next
                
    def createDLL(self,val):
        node=Node(val)
        node.prev = None
        node.next = None
        self.head =node
        self.tail=node
        
        return "The DLL is created successfullt"
    
    #insert in Doubly linked list
    def insertDll(self,val,location):
        if self.head is None:
            print("Node cannot be inserted")
        else:
            newNode = Node(val)
            
            #insert at start
            if location==0:
                newNode.prev = None
                newNode.next=self.head
                self.head.prev = newNode
                self.head=newNode
                
            #insert at the end
            elif location == -1:
                newNode.next = None
                newNode.prev = self.tail
                self.tail.next = newNode
                self.tail=newNode
            
            #insert at any index   
            else:
                tempnode = self.head
                index=0
                while index<location-1:
                    tempnode= tempnode.next
                    index+=1
                    
                nextNode = tempnode.next
                
                newNode.next = nextNode
                newNode.prev = tempnode
                nextNode.prev= newNode
                tempnode.next = newNode
                
    def traversaldl(self):
        if self.head is None:
            print("The node is not present")
        else:
            tempnode = self.head
            while tempnode:
                print(tempnode.data, end="->")
                tempnode=tempnode.next
        print()
     
    def reverseDLL(self):
        if self.head is None:
            print(" The node is not present")
        else:
            tempnode = self.tail
            while tempnode:
                print(tempnode.data,end="->")
                tempnode=tempnode.prev
            
    #delete node from DLL
    def deleteNode(self, location):
        if location==0:
            if self.head is None:
                self.head = None
                self.tail = None
                
            else:
                self.head = self.head.next
                self.head.prev = None
        elif location == -1:
            
            if self.head==self.tail:
                self.head = None
                self.tail = None
            else:
                self.tail = self.tail.prev
                self.tail.next= None
        else:
            tempnode = self.head
            index = 0
            
            while index<location-1:
                tempnode = tempnode.next
                index+=1
            
            tempnode.next = tempnode.next.next
            tempnode.next.prev = tempnode
        print("deleted the node")
            
        
                      
                
                
                
                
    
    
dll = DoublyLL()
dll.createDLL(5)
print([node.data for node in dll])

dll.insertDll(0,0)
dll.insertDll(4,-1)
dll.insertDll(3,2)
dll.insertDll(5,1)
print([node.data for node in dll])



dll.traversaldl()
dll.reverseDLL()
        
dll.deleteNode(3)
print([node.data for node in dll])
            
        
```
