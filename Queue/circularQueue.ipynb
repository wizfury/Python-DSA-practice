```python
class Queue:
    def __init__(self,maxsize):
        self.items=maxsize*[None]
        self.maxsize=maxsize
        self.start=-1
        self.top=-1
    
    def __str__(self):
        values = [str(x) for x in self.items]
        return ''.join(values)
    
    def isEmpty(self):
        if self.top==-1:
            return True
        else:
            return False
        
    def isfull(self):
        if self.top+1==self.start:
            return True
        if self.top+1==self.maxsize and self.start==0:
            return True
        else:
            return False
        
    def enqueue(self,value):
        #check  is the stack is full
        if self.isfull():
            return "The stack is full"
        else:
            #check if the top is at the end of queue
            if self.top+1==self.maxsize:
                #if yes, update it to zero
                self.top=0
            else: 
                #increase the top 
                self.top+=1
                #check if start is at -1
                if self.start == -1:
                    #set start to 0
                    self.start=0
            self.items[self.top]=value 
            return 'The element is inserted at the end of the queue' 
        
    def dequeue(self):
        #check if the queue is empty
        if self.isEmpty():
            return "The queue is empty"
        else:
            
            firstelement = self.items[self.start]
            start = self.start
            
            #check if the queue has only one element
            if self.top==self.start:
                self.start=-1
                self.top=-1
            
            #elif checko it the start is at the end
            elif self.start+1==self.maxsize:
                #move the start to beginning of the queue
                self.start=0
            
            else:
                self.start+=1
            self.items[start]=None
            return firstelement 
        
    #printing the first element in the queue   
    def peek(self):
        if self.isEmpty():
            return "The Queue is empty"
        else:
            return self.items[self.start]
        
    def delete(self):
        self.items = [None]*self.maxsize
        self.top = -1
        self.start=-1
        
        
        
    
            
            
customqueue = Queue(3)
customqueue.enqueue(1)

print(customqueue)
print(customqueue.dequeue())
print(customqueue.dequeue())
print(customqueue.dequeue())

print(customqueue.dequeue())


print(customqueue)  # Output: False

            
    
    
    
        
```
