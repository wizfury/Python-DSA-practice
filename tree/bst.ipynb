```python
from queue import Queue
class TreeNode:
    def __init__(self, x):
        self.data = x
        self.left = None
        self.right = None
        
    
newBt = TreeNode("Drinks")
leftchild = TreeNode("Hot")
rigthchild = TreeNode("Cold")
newBt.left = leftchild
newBt.right = rigthchild


#Preorder
#Root Left Right
def preorderTraversel(rootNode):
    if not rootNode:
        return
    print(rootNode.data)
    preorderTraversel(rootNode.left)
    preorderTraversel(rootNode.right)
 

#inorder 
#Left Root Right  
def inorderTraversel(rootNode):
    if not rootNode:
        return 
    inorderTraversel(rootNode.left)
    print(rootNode.data)
    inorderTraversel(rootNode.right)
    

#Postorder
#Left Right Root
def postorderTraversel(rootNode):
    if not rootNode:
        return
    postorderTraversel(rootNode.left)
    postorderTraversel(rootNode.right)
    print(rootNode.data)
    
    
def levelorderTraversal(rootNode):
    if rootNode==None:
        return
    q=Queue()
    q.put(rootNode)
    while   q.empty()==False:
        currnode = q.get()
        if currnode == None:
            continue
        print(currnode.data)
        q.put(currnode.left)
        q.put(currnode.right)  
        
def searchBT(rootnode,nodeval):
    if rootnode is None:
        return 
    q=Queue()
    q.put(rootnode)
    while q.empty()==False:
        currNode = q.get()
        if currNode.data==nodeval:
            return "success"
        q.put(currNode.left)
        q.put(currNode.right)
        
    return "Not found"

def insertNode(rootnode,newNode):
    if rootnode is None:
        return 
    q = []
    q.append(rootnode)
    while q:
        temp = q.pop(0)
        
        if temp.left is None:
            temp.left = newNode
            break
        else:
            q.append(temp.left)
        
        if temp.right is None:
            temp.right = newNode
            break
        else:
            q.append(temp.right)
    return rootnode


def deepestNode(rootNode):
    #check if the rootnode is there
    if rootNode is None:
        return
    else:
        #make a queue and insert the root
        q=Queue()
        q.put(rootNode)
        # loop until the queue is empty
        while not q.empty():
            
            root = q.get()
            
            if root.right:
                q.put(root.right)
            
            if root.left:
                q.put(root.left)
                
        deepestnode = root.data 
        return deepestnode
        
    
    
def deletedeepestNode(rootNode,dNode):
    if not rootNode:
        return
    else:
        q=Queue()
        q.put(rootNode)
        
        while not q.empty():
            root = q.get()
            
            if root.data is dNode:
                root = None
                return
            if root.left:  
                if root.left.data == dNode:
                    root.left = None
                    return
                else:
                    q.put(root.left)
            
            if root.right:
                if root.right.data is dNode:
                    root.right=None
                    return 
                else:
                    q.put(root.right) 
                    
                    
def deleteNodeBT(rootNode,node):
    if not rootNode:
        return
    else:
        q=Queue()
        q.put(rootNode)
        while not q.empty():              
            root = q.get()
            if root.data == node:
                dnode = deepestNode(rootNode)
                root.data = dnode
                deletedeepestNode(rootNode,dnode)
                return "The node has been successfully deleted"
                    
            if root.left:
                q.put(root.left)
                    
            if root.right:
                q.put(root.right)
        return "Failed to delete"


def deleteBT(rootNode):
    rootNode.data = None
    rootNode.left=None
    rootNode.right=None
    return "The BT has been successfully deleted"


    
            

          
bt = TreeNode("Drinks")
hot=TreeNode("Hot")
cold=TreeNode("cold") 
tea = TreeNode("Tea")

bt.left =hot
bt.right = cold

searchBT(bt,"Hot")
x=insertNode(bt,tea)
preorderTraversel(x)
print()
deleteBT(bt)
preorderTraversel(bt)






        


    
```
