```python
from queue import Queue
class TreeNode:
    def __init__(self, x):
        self.data = x
        self.left = None
        self.right = None
        
    
newBt = TreeNode("Drinks")
leftchild = TreeNode("Hot")
rigthchild = TreeNode("Cold")
newBt.left = leftchild
newBt.right = rigthchild


#Preorder
#Root Left Right
def preorderTraversel(rootNode):
    if not rootNode:
        return
    print(rootNode.data)
    preorderTraversel(rootNode.left)
    preorderTraversel(rootNode.right)
 

#inorder 
#Left Root Right  
def inorderTraversel(rootNode):
    if not rootNode:
        return 
    inorderTraversel(rootNode.left)
    print(rootNode.data)
    inorderTraversel(rootNode.right)
    

#Postorder
#Left Right Root
def postorderTraversel(rootNode):
    if not rootNode:
        return
    postorderTraversel(rootNode.left)
    postorderTraversel(rootNode.right)
    print(rootNode.data)
    
    
def levelorderTraversal(rootNode):
    if rootNode==None:
        return
    q=Queue()
    q.put(rootNode)
    while q.empty()==False:
        currnode = q.get()
        if currnode == None:
            continue
        print(currnode.data)
        q.put(currnode.left)
        q.put(currnode.right)  
        
def searchBT(rootnode,nodeval):
    if rootnode is None:
        return 
    q=Queue()
    q.put(rootnode)
    while q.empty()==False:
        currNode = q.get()
        if currNode.data==nodeval:
            return "success"
        q.put(currNode.left)
        q.put(currNode.right)
        
    return "Not found"

def insertNode(rootnode,newNode):
    if rootnode is None:
        return 
    q = []
    q.append(rootnode)
    while q:
        temp = q.pop(0)
        
        if temp.left is None:
            temp.left = newNode
            break
        else:
            q.append(temp.left)
        
        if temp.right is None:
            temp.right = newNode
            break
        else:
            q.append(temp.right)
    return rootnode
    
        
        
        
          
bt = TreeNode("Drinks")
hot=TreeNode("Hot")
cold=TreeNode("cold") 
tea = TreeNode("Tea")

bt.left =hot
bt.right = cold

searchBT(bt,"Hot")
x=insertNode(bt,tea)
preorderTraversel(x)
        


    
```
