```python
import queue
class BSTNode:
    def __init__(self, data):
        self.data=data
        self.left=None
        self.right=None
        
def insertNode(rootNode,nodeValue):
    #check root
    if rootNode.data==None:
        rootNode.data = nodeValue
        
    #check left
    elif nodeValue <= rootNode.data:
        if rootNode.left is None:
            rootNode.left = BSTNode(nodeValue)
        else:
            insertNode(rootNode.left,nodeValue)
    
    else:
        if rootNode.right is None:
            rootNode.right = BSTNode(nodeValue)
        else:
            insertNode(rootNode.right,nodeValue)
    return "The node has been inserted"

def preordertraverse(rootNode):
    if not rootNode:
        return 
    print(rootNode.data)
    preordertraverse(rootNode.left)
    preordertraverse(rootNode.right)
    
def inordertraverse(rootNode):
    if not rootNode:
        return 
    
    inordertraverse(rootNode.left)
    print(rootNode.data)
    inordertraverse(rootNode.right)
    
def postordertraverse(rootNode):
    if not rootNode:
        return 
    
    postordertraverse(rootNode.left)
    postordertraverse(rootNode.right)
    print(rootNode.data)

def levelordertraversal(rootNode):
    if not rootNode:
        return
    else:
        q = queue()
        q.put(rootNode)
        while not q.empty():
            root = q.get()
            print(root.data)
            
            if root.left is not None:
                q.put(root.left)
            if root.right is not None:
                q.put(root.right)     
        
def minValueNode(bsNode):
    current = bsNode
    while current.left is not None:
        current = current.left
        
    return current

def deleteNode(rootNode,nodeValue):
    if rootNode is None:
        return rootNode
    else:
        if nodeValue<rootNode.data:
            rootNode.left = deleteNode(rootNode.left,nodeValue)
        elif nodeValue>rootNode.right:
            rootNode.right = deleteNode(rootNode.right,nodeValue)
            
        else:
            if rootNode.left is None:
                temp = rootNode.right
                
                rootNode = None
                return temp
            if rootNode.right is None:
                temp = rootNode.left
                rootNode = None
                return temp
            
                
          
    
newBst = BSTNode(None)

insertNode(newBst,10)
insertNode(newBst,15)
insertNode(newBst,8)

inordertraverse(newBst)





```
